```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\juanl\Downloads\dados_vendas_sinteticos_asa.csv", encoding='latin1', sep=';', index_col="data_dia", parse_dates=True, dayfirst=True)
df
```

```python
#Filtrar para somente o interfolhado referente à equipe ItaMind:
filtro_itamind = df["Equipe responsável"] == "ItaMind: Inovação que conecta realidades e soluções"
#Aplicar filtro ao dataset:
df_itamind = df[filtro_itamind]
df_itamind
```

```python
df_itamind.describe()
```

```python
# Calcular a média com base apenas nos dados da ItaMind
desvio_padrao = df_itamind["total_venda_dia_kg"].std()
media = df_itamind["total_venda_dia_kg"].mean()

# Aplicar o filtro diretamente no df_itamind
filtro_maior_normal = df_itamind["total_venda_dia_kg"] > media + 10
df_itamind_maiores = df_itamind[filtro_maior_normal]
df_itamind_maiores
```

```python
filtro_menor_normal = df_itamind["total_venda_dia_kg"] < media - 10
df_itamind_menores = df_itamind[filtro_menor_normal]
df_itamind_menores
```

```python
import matplotlib.dates as mdates

# Resetando o índice para recuperar a coluna 'data_dia'
df_reset = df_itamind.reset_index()

# Convertendo datas para números
df_reset["data_dia_num"] = df_reset["data_dia"].apply(mdates.date2num)

# Gerar gráfico de Regressão Linear
plt.figure(figsize=(12,6))
sns.regplot(x="data_dia_num", y="total_venda_dia_kg", data=df_reset, scatter_kws={"color": "blue"}, line_kws={"color": "red"})

# Ajustando os ticks para mostrar datas legíveis
plt.xticks(ticks=df_reset["data_dia_num"][::7], labels=df_reset["data_dia"][::7].dt.strftime("%d/%m"), rotation=45)

plt.title("Regressão Linear: Vendas de Asa de Frango Congelada")
plt.xlabel("Data")
plt.ylabel("Venda total por dia (kg)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Gerar gráfico de barras para entender qual mês se superou no número de vendas
# Criar coluna com mês e ano
df_reset["mes"] = df_reset["data_dia"].dt.to_period("M")

# Agrupar por mês
vendas_mensais = df_reset.groupby("mes")["total_venda_dia_kg"].sum().reset_index()

plt.figure(figsize=(10,6))
plt.bar(vendas_mensais["mes"].astype(str), vendas_mensais["total_venda_dia_kg"], color="skyblue")

plt.title("Total de Vendas de Asa de Frango por Mês")
plt.xlabel("Mês")
plt.ylabel("Vendas (kg)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```
